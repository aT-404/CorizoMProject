```python
#Calling the data
import pandas as pd
df = pd.read_csv(r"spotify dataset.csv")
df
```

```python
#Checking for null values (Data Pre-processing)
null_count = df.isnull().sum()
null_count
```

```python
df.dropna(inplace=True) #Deleting rows with null value
#Checking for null values (Data Pre-processing)
null_count = df.isnull().sum()
null_count
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Data analysis and visualization
sample_size = int(0.03 * len(df))
random_sample1 = df.sample(n=sample_size, random_state=42)

# Scatter plot
plt.scatter(random_sample1["energy"], random_sample1["valence"])
plt.xlabel("Energy")
plt.ylabel("Valence")
plt.title("Energy vs. Valence")
plt.show()
```

```python
# Data analysis and visualization
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
# Clustering (example using K-means)
features_for_clustering = df[['danceability', 'energy', 'valence']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_for_clustering)
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Plot clusters
sns.scatterplot(x='danceability', y='energy', hue='cluster', data=df)
plt.title('Clusters based on Danceability and Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()
```

```python
# Data analysis and visualization
plt.figure(figsize=(8, 6))  # Set the figure size

# Create the histogram
plt.hist(df["danceability"], bins=20, color="skyblue", edgecolor="black")
plt.xlabel("Danceability")
plt.ylabel("Frequency")
plt.title("Distribution of Danceability")
plt.grid(True)  # Add grid lines
```

```python
# Data analysis and visualization
# Example: Histogram of track popularity
plt.hist(df['track_popularity'], bins=20)
plt.xlabel('Track Popularity')
plt.ylabel('Count')
plt.title('Distribution of Track Popularity')
plt.show()
```

```python
sample_size = int(0.05 * len(df))
random_sample2 = df.sample(n=sample_size, random_state=42)
plt.figure(figsize=(8, 6))  # Set the figure size

# Create the box plot
sns.boxplot(x="mode", y="energy", data=random_sample2 , palette="Set2")
plt.xlabel("Mode")
plt.ylabel("Energy")
plt.title("Energy Distribution by Mode")
plt.grid(True)  # Add grid lines
```
