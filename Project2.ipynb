```python
# Data analysis and visualization
sample_size = int(0.03 * len(df))
random_sample1 = df.sample(n=sample_size, random_state=42)

# Scatter plot
plt.scatter(random_sample1["energy"], random_sample1["valence"])
plt.xlabel("Energy")
plt.ylabel("Valence")
plt.title("Energy vs. Valence")
plt.show()
```

```python
# Data analysis and visualization
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
# Clustering (example using K-means)
features_for_clustering = df[['danceability', 'energy', 'valence']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_for_clustering)
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Plot clusters
sns.scatterplot(x='danceability', y='energy', hue='cluster', data=df)
plt.title('Clusters based on Danceability and Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Define your feature columns
head = ['track_popularity', 'danceability', 'energy', 'instrumentalness', 'liveness', 'duration_ms']
x = df.loc[:, head]
y = df.iloc[:, 9]

# Split data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.80)

# Define hyperparameters to search
param_grid = {
    'n_estimators': [100, 200, 300],  # Experiment with different values
    'max_depth': [None, 10, 20],  # Adjust max depth
    'min_samples_split': [2, 5, 10]  # Tune min samples split
}

# Perform grid search
grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)

# Get best parameters and score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Accuracy Score:", best_score)
```
